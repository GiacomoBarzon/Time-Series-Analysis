# Task 1

### To smooth out day-to-day fluctuations, average observations over yearly periods. Plot the first return-map from the resulting time series. What do you notice? What is the oscillation period in years?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_excel('sunspotData.xls')
```

```python
# Averaging observation over yearly periods
meanData = data.groupby(data.years).mean()

# Find local maxima
counts = np.array(meanData.sunspots)
isMax = np.r_[True, counts[1:] > counts[:-1]] & np.r_[counts[:-1] > counts[1:], True]

# Plot average data
plt.figure(figsize=[10,6])
plt.plot(meanData)
plt.plot(meanData[isMax],'o')
plt.xlabel('year', fontsize = 18)
plt.ylabel('sunspot counts', fontsize = 18)
plt.grid()
plt.title('Average sunspot count per year', fontsize = 20)

# Time btw two peaks
print('Time btw two peaks:')
print(np.diff(meanData[isMax].index))
print('Average time btw two peaks:')
print(np.mean(np.diff(meanData[isMax].index)))
```

```python
# Return map
xt = np.delete(counts, 0)
xt1 = np.delete(counts, meanData.sunspots.shape[0]-1)

# Plot the return map
plt.figure(figsize=[10,6])
plt.plot(xt, xt1)
plt.plot(xt, xt1,'o')
plt.xlabel('$x_t$', fontsize = 18)
plt.ylabel('$x_{t+1}$', fontsize = 18)
plt.title('First return map', fontsize = 20)
plt.grid()
```

It can be easily noticed that there are some visible and almost periodic peaks of the sunspot counts over the years, with an average time period of 9 years.
The periodicity of the sunspot activity can also be seen by looking to the ciclic behaviour of the return map.

# Task 2

### 1) By using linear regression, remove the trend from the data

```python
investments = pd.read_excel('investment.xls')
investments.describe()
```

```python
# Linear regression
from sklearn.linear_model import LinearRegression

quarters = np.array(investments.index)
y = np.array(investments['investment'])

model = LinearRegression().fit(quarters.reshape(168,1), y)
```

```python
# Plot average data
plt.figure(figsize=[10,6])
plt.plot(investments)
plt.plot(quarters,model.predict(quarters.reshape(168,1)))
plt.xlabel('quarter', fontsize = 18)
plt.ylabel('investment per capita', fontsize = 18)
plt.grid()
plt.title('Scaled quarterly U.S private investment per capita rates', fontsize = 20)
```

```python
# Removing the linear trend from the data
resInvestments = y - model.predict(quarters.reshape(168,1))

plt.figure(figsize=[10,6])
plt.plot(quarters,resInvestments)
plt.xlabel('quarter', fontsize = 18)
plt.ylabel('residual investment', fontsize = 18)
plt.grid()
plt.title('Residual private investment per capita rates', fontsize = 20)
```

### b) Examine whether the residual time series is approximately stationary. Is the time series of first differences stationary? How about the time series of second-order differences?

```python
# Moving average
def moving_average(y,N):
    cumsum, moving_aves = [0], []
    n = []

    for i, x in enumerate(y, 1):
        cumsum.append(cumsum[i-1] + x)
        if i>=N:
            moving_ave = (cumsum[i] - cumsum[i-N])/N
            #can do stuff with moving_ave here
            moving_aves.append(moving_ave)
            n.append(i-N+1)
    return (n, moving_aves)
```

```python
# Trend removed moving average
N = 40
(n, moving_aves) = moving_average(resInvestments, N)

mean = np.mean(resInvestments)
dev = np.std(resInvestments)

# Plot moving average
plt.figure(figsize=[10,6])
plt.plot(n,moving_aves)
plt.hlines(mean,n[0],n[-1],linestyles='dashdot',label='mean')
plt.hlines(mean+dev,n[0],n[-1],linestyles='dashdot', label='mean+std', colors='r')
plt.hlines(mean-dev,n[0],n[-1],linestyles='dashdot', label='mean-std', colors='b')
plt.xlabel('sliding window', fontsize = 18)
plt.ylabel('average', fontsize = 18)
plt.grid()
plt.legend()
plt.title('Moving averages with sliding windows of length = %i' %N, fontsize = 20)
```

```python
# First differences series
first_x = np.delete(quarters,0)
first_diff_inv = np.diff(y)

# Plot
plt.figure(figsize=[10,6])
plt.plot(first_x,first_diff_inv)
plt.xlabel('quarter', fontsize = 18)
plt.ylabel('1st diff investment', fontsize = 18)
plt.grid()
plt.title('First difference series of investment per capita rates', fontsize = 20)

# Trend removed moving average
N = 40
(n, moving_aves) = moving_average(first_diff_inv, N)

mean = np.mean(first_diff_inv)
dev = np.std(first_diff_inv)

# Plot moving average
plt.figure(figsize=[10,6])
plt.plot(n,moving_aves)
plt.hlines(mean,n[0],n[-1],linestyles='dashdot',label='mean')
plt.hlines(mean+dev,n[0],n[-1],linestyles='dashdot', label='mean+std', colors='r')
plt.hlines(mean-dev,n[0],n[-1],linestyles='dashdot', label='mean-std', colors='b')
plt.xlabel('sliding window', fontsize = 18)
plt.ylabel('average', fontsize = 18)
plt.grid()
plt.legend()
plt.title('Moving averages with sliding windows of length = %i' %N, fontsize = 20)
```

```python
# Second order differences series
second_x = np.delete(first_x,0)
second_diff_inv = np.diff(y, n=2)

# Plot
plt.figure(figsize=[10,6])
plt.plot(second_x,second_diff_inv)
plt.xlabel('quarter', fontsize = 18)
plt.ylabel('2nd diff investment', fontsize = 18)
plt.grid()
plt.title('Second difference series of investment per capita rates', fontsize = 20)

# Trend removed moving average
N = 40
(n, moving_aves) = moving_average(second_diff_inv, N)

mean = np.mean(second_diff_inv)
dev = np.std(second_diff_inv)

# Plot moving average
plt.figure(figsize=[10,6])
plt.plot(n,moving_aves)
plt.hlines(mean,n[0],n[-1],linestyles='dashdot',label='mean')
plt.hlines(mean+dev,n[0],n[-1],linestyles='dashdot', label='mean+std', colors='r')
plt.hlines(mean-dev,n[0],n[-1],linestyles='dashdot', label='mean-std', colors='b')
plt.xlabel('sliding window', fontsize = 18)
plt.ylabel('average', fontsize = 18)
plt.grid()
plt.legend()
plt.title('Moving averages with sliding windows of length = %i' %N, fontsize = 20)
```

By looking at the plots of the moving-average of the three series (without linear trend, first differences, second differences), it can be noticed that the all the obtained values are close to the average of the whole sample (contained in the interval $[\overline{x}-\sigma, \overline{x}+\sigma]$) and there aren't systematic trend across time: this shows that the average is almost constant in time and it's a reason why the series can be considered stationary.

### c) Compute the autocorrelation function of the time series with the linear trend removed. Are business cycles (corresponding to peaks in the autocorrelation function) periodic?

The time series is assumed to be stationary and ergodic, so the empirical estimation of the auto-covariance and auto-correlation are computed as:<br/>
$acov(\Delta t) = \gamma(\Delta t) = \dfrac{1}{T-\Delta t} \sum_{i=1}^{T-\Delta t}(x_t-\overline{x})(x_{t+\Delta t}-\overline{x})$ <br/>
$acorr(\Delta t) = \rho(\Delta t) = \dfrac{\gamma (\Delta t)}{\gamma (0)}$

```python
def acov(x, dt):
    T = len(x)
    x_mean = np.mean(x)
    x = x - x_mean
    
    cov = 0
    
    # maybe try convolution function
    for i in range(T-dt):
        cov += x[i]*x[i+dt]

    return cov / (T-dt)
        

def acorr(x, dt):
    return acov(x,dt) / acov(x,0)
```

```python
# Compute autocorrelation function
corr_inv = np.zeros(len(quarters)-1)

for i in range(len(quarters)-1):
    corr_inv[i] = acorr(resInvestments, quarters[i])
    
# Find local maxima
isMax = np.r_[True, corr_inv[1:] > corr_inv[:-1]] & np.r_[corr_inv[:-1] > corr_inv[1:], True]
    
# Plot
plt.figure(figsize=[10,6])
plt.plot(quarters[:-1],corr_inv)
#plt.plot(quarters[:-1],corr_inv,'.')
plt.plot(quarters[:-1][isMax],corr_inv[isMax],'o')
plt.xlabel('$\Delta t$', fontsize = 18)
plt.ylabel(r'$\rho (\Delta t)$', fontsize = 18)
plt.grid()
plt.title('Autocorrelation functions', fontsize = 20)

# Time btw two peaks
print('Time btw two peaks:')
print(np.diff(quarters[:-1][isMax]))
print('Average time btw two peaks:')
print(np.mean(np.diff(quarters[:-1][isMax])))
```

The autocorrelation functions has many peaks, that correspond to business cycles. As it can be noticed above, the values of time between two peaks is very variable: this means that the business cycles are not periodic.

# Task 3

### 1) Create your own AR time series of length T = 200 and order p = 4, with the following coefficients given:
### a0 = 0.0, a1 = −0.8, a2 = 0.0, a3 = 0.0, and a4 = 0.4, with $\epsilon_t$ ∼ N(0,1), and with the initial value of the time series being $x_0$ = 0.0

AR series: $x_t = a_0 + \sum_{i=1}^{p} a_i x_{t-1} + \epsilon_t $

```python
# Parameters
T = 200
p = 4

# Coefficients
a0 = 0.0
a1 = -0.8
a2 = 0.0
a3 = 0.0
a4 = 0.4

a = np.array([a1, a2, a3, a4])

# Initial value
x0 = 0.0
```

```python
# Noise
seed = 30091996
np.random.seed(seed)
epsilon = np.random.normal(size=T)
```

```python
# Generate AR series
x = np.zeros(T+1)
x[0] = x0          # setting initial value

for t in range(1,T+1):
    temp = 0
    for i in range(0,p):
        if(t-i-1 > 0):
            temp += a[i]*x[t-i-1]
    x[t] = a0 + temp + epsilon[t-1]
```

### 2) Plot the time series in time as well as the first return-map. What do you notice?

```python
# Plot the time series
plt.figure(figsize=[10,6])
plt.plot(np.arange(0,T+1),x)
#plt.plot(quarters[:-1],corr_inv,'.')
plt.xlabel('t', fontsize = 18)
plt.ylabel(r'$x_t$', fontsize = 18)
plt.grid()
plt.title('AR(p=%i) time series' %p, fontsize = 20)
```

```python
# Return map
xt = np.delete(x, 0)
xt1 = np.delete(x, len(x)-1)

# Plot the return map
plt.figure(figsize=[10,6])
plt.plot(xt, xt1)
plt.plot(xt, xt1,'o')
plt.xlabel('$x_t$', fontsize = 18)
plt.ylabel('$x_{t+1}$', fontsize = 18)
plt.grid()
plt.title('First return map', fontsize = 20)
```

It can be noticed that the trend of the time series seems to be an exponential with values that continuously jump from negative to positive and the other way around. This is reflected in the plot of the first return map, where the points are distributed along a line with negative slop.

### We have derived that the criterion for an AR(1) process, $x_t = a_0 + a_1x_{t−1} + ε_t$, $\epsilon_t ∼ W(0,\sigma^2)$, to be stationary is |a1|<1. Show that for an AR(p) process, p>1, the mean of the time series is stationary if |ai|<1, by expanding the process in time

By expanding the process in time I obtain:<br/>
$x_t = a_0 + \sum_{i=1}^{p}a_i x_{t-i} + \epsilon_t$ <br/>
$    = a_0 + \sum_{i=1}^{p}a_i \{a_0 + \sum_{j=1}^{p}a_j x_{t-i-j} + \epsilon_{t-1}\} + \epsilon_t$ <br/>
$    = a_0 + \sum_{i=1}^{p}a_i \{a_0 + \sum_{j=1}^{p}a_j [a_0 + \sum_{k=1}^{p}a_k x_{t-i-j-k} + \epsilon_{t-2}] + \epsilon_{t-1}\} + \epsilon_t$ <br/>

By calculating the mean value, all the term linear in $\epsilon_t$ vanishes, because of the definition of white noise ($E[\epsilon_t]=\overline{\epsilon}_t=0 \ \forall t)$, so:

$\overline{x}_t = a_0 + \sum_{i=1}^{p}a_i \overline{x}_{t-i} + \overline{\epsilon}_t = a_0 + \sum_{i=1}^{p}a_i \overline{x}_{t-i}$ <br/>
$ = a_0 + \sum_{i=1}^{p}a_i \{a_0 + \sum_{j=1}^{p}a_j \overline{x}_{t-i-j}\} $ <br/>
$    = a_0 + \sum_{i=1}^{p}a_i \{a_0 + \sum_{j=1}^{p}a_j [a_0 + \sum_{k=1}^{p}a_k \overline{x}_{t-i-j-k}] \}$ <br/>
$ = a_0 + a_0\sum_{i=1}^{p} a_i + a_0\sum_{i=1}^{p} a_i \sum_{j=1}^{p} a_j + \sum_{i=1}^{p} a_i \sum_{j=1}^{p} a_j \sum_{k=1}^{p} a_k(a_0 + ...)$ <br/>
$ = a_0 * [1 + \sum_{i=1}^{p} a_i + (\sum_{i=1}^{p} a_i)^2 + ... + (\sum_{i=1}^{p} a_i)^{T-p} + ...] $ <br/>

The first (T-p) terms in the summation can be rewritten as: <br/>
$a_0 \sum_{n=0}^{T-p}(\sum_{i=1}^{p} a_i)^n $ <br/>
while the other term depends on the first (finite) p values of the AR process. <br/>

By taking the limit for $T\to 0$ and using the property of the geometric series the expectation value becomes: <br/>
$\overline{x}_t = a_0 \sum_{n=0}^{\infty}(\sum_{i=1}^{p} a_i)^n = \dfrac{a_0}{1 - \sum_{i=1}^{p}a_i}$ if $\vert \sum_{i=1}^{p}a_i \vert < 1$
