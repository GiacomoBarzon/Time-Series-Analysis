## Task 1: AR(p) models

```python
#load .mat file
mat_file = loadmat('Tut2_file1.mat')
data = mat_file["x"][:]
data = np.array(data).reshape(data.size,1)
print(data.size)
```

### 1) Estimate coefficients of AR(1)  

In order to find the coefficients $a_0$ and $a_1$ of an AR(1) model it is possible to perform a linear regression on data using $x_{t-1}$ as x and $x_t$ as y, since an AR(1) process reads as $x_t=a_0+a_1 x_{t-1} + \epsilon_t$.

```python
from sklearn.linear_model import LinearRegression

# Perform linear regression on data
reg = LinearRegression().fit(data[0:len(data)-1], data[1:len(data)])
print('Linear model: \n')
print('a_0 = ', reg.intercept_,' , a_1 = ', reg.coef_)
```

```python
# Plot results

t = np.arange(0, len(data), 1)

fig1 = plt.subplots(figsize=[15,10])

plt.subplot(1,2,1)
plt.plot(t, data, '-b')
plt.title("Plot of the Time Series", fontsize = 15)
plt.xlabel("t", fontsize = 15)
plt.ylabel("$x_t$", fontsize = 15)

plt.subplot(1,2,2)
plt.plot(data[0:len(data)-1], data[1:len(data)], 'o', label = 'Return plot')
plt.plot(data[0:len(data)-1], reg.coef_*data[0:len(data)-1]+reg.intercept_, '--r', label = 'Linear Fit')
plt.title("Return Plot and linear fit", fontsize = 15)
plt.xlabel("$x_{t-1}$", fontsize = 15)
plt.ylabel("$x_t$", fontsize = 15)
plt.legend()

plt.show(fig1)
```

There is clearly a trend in the original Time Series; in particular a linear trend that is the same appearing in the Return Plot. This suggests that the coefficients $a_0$ and $a_1$ just found are actually influenced by the trend in the original series. In particular $a_1\sim 1$, i.e. the value of $x_t$ is nearly that of a random walk, plus a systematic drift given by the value of $a_0$. This means that the process has a drift on which acts the white noise, so it is necessary to regress out the series:

### 2)  Detrend Time Series

```python
# Evaluate trend and subtract from the series 

reg2 = LinearRegression().fit(t.reshape(-1,1), data)
print('Detrended Time Series: \n')
print('Angular coefficient = ', reg2.coef_,' , Intercept = ', reg2.intercept_)
```

```python
new_data = data - (reg2.coef_*t.reshape(-1,1) + reg2.intercept_)

reg3 = LinearRegression().fit(new_data[0:len(new_data)-1], new_data[1:len(new_data)])
print('Linear model on new data: \n')
print('a_0 = ', reg3.intercept_,' , a_1 = ', reg3.coef_)
```

```python
# Plot results

fig2 = plt.subplots(figsize=[15,10])

plt.subplot(1,2,1)
plt.plot(t, new_data, '-b')
plt.title("Plot of the new Time Series", fontsize = 15)
plt.xlabel("t", fontsize = 15)
plt.ylabel("$x_t$", fontsize = 15)

plt.subplot(1,2,2)
plt.plot(new_data[0:len(new_data)-1], new_data[1:len(new_data)], 'o', label = 'Return plot')
plt.plot(new_data[0:len(new_data)-1], reg3.coef_*new_data[0:len(new_data)-1]+reg3.intercept_, '--r', label = 'Linear Fit')
plt.title("Return Plot and linear fit", fontsize = 15)
plt.xlabel("$x_{t-1}$", fontsize = 15)
plt.ylabel("$x_t$", fontsize = 15)
plt.legend()

plt.show(fig2)
```

Now the series seems much more noisy and without a trend. This can be seen also in the Return Plot, where now the points form a sort of blob in the center. Now the $a_1$ coefficient is much less than 1 (so the series is stationary), even if it seems not to capture completely the nature of the series, because it is not on the axis of the ellipse. Nevertheless now the linear model is really doing better than before. This can be checked through a prediction of values based on the linear model just found:

### 3)  Generate predictions from estimated AR(1) model

```python
# Generate five steps ahead
a0 = reg3.intercept_
a1 = reg3.coef_

new_data = new_data.reshape(len(new_data))
 
prediction = np.zeros((len(new_data),5))      # Generate vector of predictions

prediction[:len(new_data)-1,0] = a0 + a1*new_data[:len(new_data)-1]
prediction[:len(prediction)-1,1] = a0 + a1*prediction[0:len(prediction)-1,0]
prediction[:len(prediction)-2,2] = a0 + a1*prediction[0:len(prediction)-2,1]
prediction[:len(prediction)-3,3] = a0 + a1*prediction[0:len(prediction)-3,2]
prediction[:len(prediction)-4,4] = a0 + a1*prediction[0:len(prediction)-4,3]

# Now compare prediction with true values and compute errors
errors = np.zeros((len(new_data),5))

errors[:len(prediction)-1,0] = abs(prediction[:len(prediction)-1,0] - new_data[1:])
errors[:len(prediction)-2,1] = abs(prediction[:len(prediction)-2,1] - new_data[2:])
errors[:len(prediction)-3,2] = abs(prediction[:len(prediction)-3,2] - new_data[3:])
errors[:len(prediction)-4,3] = abs(prediction[:len(prediction)-4,3] - new_data[4:])
errors[:len(prediction)-5,4] = abs(prediction[:len(prediction)-5,4] - new_data[5:])

# Print errors on generate values
print('Error on first prediction: ', np.mean(errors[:,0]), ' +- ', np.std(errors[:,0]))
print('Error on second prediction: ', np.mean(errors[:,1]), ' +- ', np.std(errors[:,1]))
print('Error on third prediction: ', np.mean(errors[:,2]), ' +- ', np.std(errors[:,2]))
print('Error on fourth prediction: ', np.mean(errors[:,3]), ' +- ', np.std(errors[:,3]))
print('Error on fifth prediction: ', np.mean(errors[:,4]), ' +- ', np.std(errors[:,4]))
```

Here are shown the absolute values of the errors on the various predictions along with their standard deviations. As one can expect, the error grows with the increasing order of the prediction: in particular there is a neat increase from the first to the second prediction and then it more or less stabilizes, as the increase in error decreases along with increasing step; also the standard deviation of the errors increases as the order of prediction increases, but again not linearly. This is understandable since the initial error is given probably by the presence of the noise term, not accounted in the linear model; then the initial error propagates from step to step, but in new predictions the noise "helps" to keep the error pretty small and prevents it from growing linearly.

```python
# Define autocorrelation function (only for first n values)
def acorr(x,y,n):
    # First calculate mean and std of the TS
    mean = np.mean(y)
    std = np.std(y)
    
    ac = np.zeros(n)
    
    for Dt in range(0, n):
        sum = 0
        for i in range(0, int(x[-1]-Dt)):
            sum += (y[i]-mean)*(y[i+Dt]-mean)/(x[-1]-Dt)
        ac[Dt] = sum/(std**2)
    return ac
```

```python
data2_acorr = acorr(t2, data2, 1000)
```

```python
plt.figure(figsize=[12,8])
plt.plot(t2[:1000], data2_acorr, '-b')
plt.plot(t2[:1000], data2_acorr, 'ko', markersize = 5)
plt.hlines(0., 0, 1000, 'k', linestyles = 'dashed')
plt.title("Autocorrelation function", fontsize = 15)
plt.xlabel("$\Delta t$", fontsize = 15)
plt.ylabel("$acorr(\Delta t)$", fontsize = 15)

plt.show()
```

The autocorrelation function goes to 0 with increasing lag (here are shown only the first 1000 lags for simplicity), first in a monotonic way and then oscillating around 0 in what seems to be only noise. The function does not present remarkable peaks, indicating that the series is stationary but not periodic.

## Task 2: Autocorrelation and partial autocorrelation

```python
#load .mat file
mat_file_3 = loadmat('Tut2_file3.mat')
xAR = mat_file_3["x_AR"][:,:]
xMA = mat_file_3["x_MA"][:,:]

xAR = np.array(xAR).reshape(xAR.size, 1)
xMA = np.array(xMA).reshape(xMA.size, 1)
print(xAR.shape)
```

```python
# Plot of the series
t = np.arange(0, len(xAR), 1).reshape(-1,1)      # "time" vector 

fig_model_1 = plt.subplots(figsize=[15,8])

plt.subplot(1,2,1)
plt.plot(t[:500], xAR[:500], '-b')
plt.title('First 500 samples of AR series', fontsize=15)
plt.xlabel('t', fontsize=15)
plt.ylabel('$x_t$', fontsize=15)

plt.subplot(1,2,2)
plt.plot(t[:500], xMA[:500], '-b')
plt.title('First 500 samples of MA series', fontsize=15)
plt.xlabel('t', fontsize=15)
plt.ylabel('$x_t$', fontsize=15)

plt.show(fig_model_1)
```

```python
# Define partial autocorrelation function (only for first n values)
def pacorr(x,y,n):
    
    pac = np.zeros(n)
    
    pac[0] = 1.
    pac[1] = acorr(x,y,n)[1]
    
    for Dt in range(2, n):
        data = np.zeros((len(y)-Dt+1, Dt-1))
        data2 = np.zeros((len(y)-Dt+1, Dt-1))
        for i in range (1, Dt):
            data[:,i-1] = y[Dt-i-1:len(y)-i,0]
            data2[:,i-1] = y[i:len(y)-Dt+i+1,0]
        reg1 = LinearRegression().fit(data,y[Dt-1:,0])
        reg2 = LinearRegression().fit(data2,y[:len(y)-Dt+1,0])
        true_data1 = y[Dt-1:,0] - (np.dot(data, reg1.coef_) + reg1.intercept_)
        true_data2 = y[:len(y)-Dt+1,0] - (np.dot(data2, reg2.coef_) + reg2.intercept_)
        summ = 0
        mean1 = np.mean(true_data1)
        std1 = np.std(true_data1)
        mean2 = np.mean(true_data2)
        std2 = np.std(true_data2)
        for j in range(0, int(x[-1]-Dt)):
            summ += (true_data2[j]-mean2)*(true_data1[j]-mean1)/(x[-1]-Dt)
        pac[Dt] = summ/(std1*std2)
        
    return pac
    
```

```python
n = 15   # number of iterations

# Auto-correlation and Partial AC for Auto-regressive series 
xAR_acorr = acorr(t, xAR, n)
xAR_pacorr = pacorr(t, xAR, n)

# Auto-correlation and Partial AC for Moving-average series 
xMA_acorr = acorr(t, xMA, n)
xMA_pacorr = pacorr(t, xMA, n)
```

```python
# Plot the results

fig_model = plt.subplots(figsize=[18,15])

plt.subplot(2,2,1)
plt.plot(t[0:n], xAR_acorr, '-b')
plt.plot(t[0:n], xAR_acorr, 'ko')
plt.title('Auto-correlation AR', fontsize=15)
plt.xlabel('$\Delta t$', fontsize=15)
plt.ylabel('acorr', fontsize=15)

plt.subplot(2,2,2)
plt.plot(t[0:n], xAR_pacorr, '-r')
plt.plot(t[0:n], xAR_pacorr, 'ko')
plt.plot(t[6], xAR_pacorr[6], 'b*', markersize=20)
plt.title('Partial Auto-correlation AR', fontsize=15)
plt.xlabel('$\Delta t$', fontsize=15)
plt.ylabel('pacorr', fontsize=15)

plt.subplot(2,2,3)
plt.plot(t[0:n], xMA_acorr, '-m')
plt.plot(t[0:n], xMA_acorr, 'ko')
plt.plot(t[5], xMA_acorr[5], 'b*', markersize=20)
plt.title('Auto-correlation MA', fontsize=15)
plt.xlabel('$\Delta t$', fontsize=15)
plt.ylabel('acorr', fontsize=15)

plt.subplot(2,2,4)
plt.plot(t[0:n], xMA_pacorr, '-g')
plt.plot(t[0:n], xMA_pacorr, 'ko')
plt.title('Partial Auto-correlation MA', fontsize=15)
plt.xlabel('$\Delta t$', fontsize=15)
plt.ylabel('pacorr', fontsize=15)

plt.show(fig_model)
```

Here the blue star shows the last non-zero value of the partial auto-correlation function in the case of the AR series and of the auto-correlation function in the case of the MA series. These values represent the order of the series, so $\textbf{p = 6 for AR}$ and $\textbf{q = 5 for MA}$.
