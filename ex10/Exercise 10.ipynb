# Time series prediction with RNN and LSTM networks in PyTorch

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# show Lorenz data
lorenz = np.load('data.npy')

fig1 = plt.subplots(figsize=[14,10])

# plot data
plt.subplot(2,1,1)
plt.plot(lorenz)
plt.title('Experimental data', fontsize = 16)
plt.xlabel('t', fontsize = 12)
plt.ylabel(r'$x_t$', fontsize = 12)
plt.grid()
    
# plot first return plot
plt.subplot(2,1,2)
plt.plot(lorenz[:-1], lorenz[1:], lw=0.3)
plt.xlabel(r'$x_{t-1}$', fontsize = 12)
plt.ylabel(r'$x_t$', fontsize = 12)
plt.title('First return plot', fontsize = 16)
plt.grid()
    
plt.subplots_adjust(hspace = 0.3)
plt.show(fig1)
```
